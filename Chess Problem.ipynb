```python
'''Here I have set up the code for n = 5. I am using duality so I check all the combinations of 3 positions on the 5x5
by placing queens in those 3 places and checking the number of pawns that can be placed safely. Below it shows that if 
I place 3 queens, max 5 pawns can be placed and so we can say that if 5 queens are there, 3 pawns can be placed safely.
The code can be tweaked for n = 6 but is pretty slow as I have not made use of symmetries yet. '''

from itertools import combinations
n = 5
board = [[] for x in range(n)]

def initialise_board(n):
    for x in range(n):
        for y in range(n):
            board[x].append(' ')
            
def reset_board(n):
    for x in range(n):
        for y in range(n):
            board[x][y] = ' '
            
def print_board():
    printy = ""
    for x in range(n):
        for y in range(n):
            if y < (n-1):
                printy = printy + board[x][y] + "|"
            else:
                printy = printy + board[x][y] + '\n'
                 
        if x < (n - 1):
            printy = printy + '-'*(n*2-1) + '\n' 
    return print(printy)


help_listy = list(range(0,n-2))

listy = list(combinations(help_listy,2))
for x in help_listy:
    listy.append((x,x))

print("listy" + listy)
    
comb = list(combinations(listy,3))

print(comb)






def f1fill_diagonal(pos):
    i1 = int(pos[0])
    i2 = int(pos[1])
    if i1 == 4 or i2 == 4:
        no = 1
    elif i1 == 3 or i2 == 3:
        no = 2
    elif i1 == 2 or i2 == 2:
        no = 3
    elif i1 == 1 or i2 == 1:
        no = 4
    else: 
        no = 5
    for x in range(no):
        board[i1][i2] = 'D'
        i1 += 1
        i2 += 1
        
def f2fill_diagonal(pos):
    i1 = int(pos[0])
    i2 = int(pos[1])
    if i1 == 0 or i2 == 0:
        no = 1
    elif i1 == 1 or i2 == 1:
        no = 2
    elif i1 == 2 or i2 == 2:
        no = 3
    elif i1 == 3 or i2 == 3:
        no = 4
    else: 
        no = 5
    for x in range(no):
        board[i1][i2] = 'D'
        i1 -= 1
        i2 -= 1
    
def b1fill_diagonal(pos):
    i1 = int(pos[0])
    i2 = int(pos[1])
    if i1 == 4 or i2 == 0:
        no = 1
    elif i1 == 3 or i2 == 1 :
        no = 2
    elif i1 == 2 or i2 == 2:
        no = 3
    elif i1 == 1 or i2 == 3:
        no = 4
    else: 
        no = 5
    for x in range(no):
        board[i1][i2] = 'D'
        i1 += 1
        i2 -= 1
        
        
def b2fill_diagonal(pos):
    i1 = int(pos[0])
    i2 = int(pos[1])
    if i1 == 0 or i2 == 4:
        no = 1
    elif i1 == 1 or i2 == 3:
        no = 2
    elif i1 == 2 or i2 == 2:
        no = 3
    elif i1 == 3 or i2 == 1:
        no = 4
    else: 
        no = 5
    for x in range(no):
        board[i1][i2] = 'D'
        i1 -= 1
        i2 += 1
        
def check(pos):
    'vertical'
    for x in range(n):
        idx2 = int(pos[1])
        board[x][idx2] = 'D'
    'horizontal'
    for x in range(n):
        idx1 = int(pos[0])
        board[idx1][x] = 'D'
    'diagonal1'
    f1fill_diagonal(pos)
    'diagonal2'
    f2fill_diagonal(pos)
    b1fill_diagonal(pos)
    b2fill_diagonal(pos)
    
def pawns():
    count = 0
    for x in range(n):
        for y in range(n):
            if board[x][y] != 'D':
                count += 1
    return count
            
        

list3 =[]
initialise_board(5)
for y in comb:
    reset_board(n)
    for x in y:
        check(x)
        print_board()
    print(pawns())    
    list3.append(pawns())

max(list3)
    
```
